Looking at the data, we decide that the best way to represent it is with a dictionary. We decide that we want to make it easy to access entries by using a combination of the model year and car name, so that will be the key. Each key-value pair in the dictionary will then consist of:
<ul>
   <li>a key that is a string beginning with the car's model year (a 2-digit string), a blank, and the car's name: for example, <code>'70 peugeot 504'</code>. We have to add in the year to the key because the data file contains data for the same car name in different years. Without the year the key wouldn't be unique.
   <li>a value that is a dictionary. That dictionary contains all of the other data (its attributes) about the car. Each key-value pair in this dictionary consists of:
   <ul>
      <li>the key is one of the column names: <code>'mpg','cyl','dsp','hp','lbs','tth'</code> and <code>'org'</code>.
      <li>the value is the data from the file. All of the data are <code>float</code> except for <code>'cyl'</code> and <code>'org'</code>, which are <code>int</code>s.
   </ul>
</ul>

You can think of this dictionary of dictionaries as a labeled table. The labels of the rows are the model year and car name; the labels of the columns are the attributes: <code>'mpg','cyl','dsp','hp','lbs','tth'</code> and <code>'org'</code>.

The first line of the file looks like:

<code>18.0   8   307.0      130.0      3504.      12.0   70  1 chevrolet chevelle malibu</code>

The entry in the dictionary that corresponds to that line should look like:

<code>{'70 chevrolet chevelle malibu': {'mpg':18.0, 'cyl':8, 'dsp':307.0, 'hp':130.0, 'lbs':3504.0, 'tth':12.0, 'org':1}}</code>.

It's a bit of a surprise that General Motors could only get 130 horsepower from a 307 cubic-inch V8. I think that this may be an error.

Phase 1 of this assignment is to read the file and create the dictionary of cars. First let's define a function <code>addCar(c, ln)</code> whose arguments will be the dictionary of cars and a line from the data file. The function will modify the car dictionary by adding a new entry into it. 

```python
def addCar(c, ln):
    # Breaks the line ln into fields: adds an entry to a dictionary of dictionaries.
    # The key consists of the year and car name; the value is the newly-created dictionary of attributes.
    # As read, a line contains a newline character at the end -- strip it off with rstrip()
    fields = ln.rstrip().split(maxsplit=8) # don't split the last field: it's the car name and has blanks
    # Now you have a list of nine fields. Field 0 is the miles-per-gallon field.
    # Your code for the computation of the key, which is the year and car name, follows:
    name = str(fields[6] + " " + fields[8])
    # Now your code should create a dictionary containing the data for mpg, cyl, dsp, hp, lbs, tth and org.
    # Don't forget to conver9t all of the fields to the appropriate numeric data type.
    car = {}
    car[name] = {'mpg':float(fields[0]), 'cyl':int(fields[1]), 'dsp':float(fields[2]), 
                 'hp':float(fields[3]), 'lbs':float(fields[4]), 'tth':float(fields[5]),
                 'org':int((fields[7]))}
    # Now your code should update c by adding a new key-value pair. There is no need of a return statement in this
    # function because it modifies its first (mutable) argument.
    c.update(car)    
```

```python
# Don't change this cell -- it's a code check.
_ = {}
addCar(_,'18.0  8  306.0  227.0  4102.  12.0  71  2  super Yugo coupe')
assert _ and _['71 super Yugo coupe']['lbs'] == 4102. 
```

Now let's define a function <code>getCars(filename)</code> whose argument will be a string that contains the qualified name (filename.filetype) of the file of data. The function creates an empty dictionary and then reads the file a line at a time, each time calling <code>addCar(cars, line)</code> to add a new entry into the dictionary. The return value of the function is the newly-created and populated dictionary of cars.

```python
# Your code to define the function getCars(filename) which creates and populates the dictionary goes here.
def getCars(filename):
    keyCars = {}
    #line = filename.readlines()    
    with open (filename, 'r') as file:
        for line in file: 
            addCar(keyCars,line)
    return keyCars
```

Now create the dictionary for all of the cars by calling <code>getCars(filename)</code> with the filename of the data file, <code>'OldCars.txt'</code>. 

```python
# Your code to create the dictionary goes here. Note that the file 'OldCars.txt' must be in the same folder as the
# .ipynb file for this notebook.
#with open ('OldCars.txt', 'r') as file:
cars = {}
cars = getCars('OldCars.txt')
    
```

```python
# Code check block - don't change this cell.
assert len(cars.keys()) == 390
assert 7 == len(cars['70 chevrolet chevelle malibu'].keys())
assert cars['70 chevrolet chevelle malibu']['lbs'] == 3504.
# Fun with string.format() :-)
def printTable(c, n=10):  # prints the first n entries in c AS A TABLE with row and column labels.
    w0 = max(len(x) for x in c)  # will be the length of the longest key
    widths = {'w0':w0 + 1, 'w1':7, 'w2':5, 'w3':5, 'w4':6, 'w5':7, 'w6':7, 'w7':5}
    hdrf = '{0:^{w0}s} |{1:^{w1}s}|{2:^{w2}s}|{3:^{w3}s}|{4:^{w4}s}|{5:^{w5}s}|{6:^{w6}s}|{7:^{w7}s}'
    print(hdrf.format('year, make and model', 'mpg', 'cyl', 'dsp', 'hp', 'lbs', 'tth', 'org', **widths))
    sepf = '{0:-^{w0}s}-|{1:-^{w1}s}|{2:-^{w2}s}|{3:-^{w3}s}|{4:-^{w4}s}|{5:-^{w5}s}|{6:-^{w6}s}|{7:-^{w7}s}'
    print(sepf.format('','','','','','','','',**widths))
    rowf = '{0:>{w0}s} |{1:^{w1}.1f}|{2:^{w2}d}|{3:^{w3}.0f}|{4:^{w4}.0f}|{5:^{w5}.0f}|{6:^{w6}.1f}|{7:^{w7}d}'
    keylist = list(c.keys())[:n]  # the first n entries
    for k in keylist:
        ats = c[k]
        att = [k, ats['mpg'], ats['cyl'], ats['dsp'], ats['hp'], ats['lbs'], ats['tth'], ats['org']]
        print(rowf.format(*att, **widths))
    
printTable(cars)
```

#### The 1970 Plymouth Barracuda

You might guess that the higher the power-to-weight ratio is, the faster the acceleration (fast acceleration means a small value of the <code>tth</code> field). We are going to find out by visualizing the data.

Compute two lists: <code>time_to_100</code>, whose members are the <code>'tth'</code> attributes of every car, and <code>power_to_weight_ratios</code>, whose members are the <code>'hp'</code> attribute divided by the <code>'lbs'</code> attribute of every car. Then we'll use the plotting code below to plot the data.

```python
# Your code to compute the tth list and the power_to_weight_ratio list goes here.
time_to_100 = []
power_to_weight_ratios = []
for c in cars :
    time_to_100.append(cars[c]['tth'])
    power_to_weight_ratios.append(cars[c]['hp']/cars[c]['lbs'])
```

```python
# Plotting code (scatter plot). Don't change this code.
import pylab as pl
%matplotlib inline
pl.plot(power_to_weight_ratios, time_to_100, 'ro')
pl.title('Does performance vary with power-to-weight ratio?')
pl.ylabel('time to 100 mph')
pl.ylim(0,26)
pl.xlabel('power/weight')
pl.xlim(0.01,0.08)
pl.show()
```

Can you spot the 'cuda in the plot? Does the plot have any outliers? What was the outlier (hint: it was the car with the largest power-to-weight ratio)?

```python
# Your code to find the car with the largest power-to-weight ratio goes here.
largest = max(power_to_weight_ratios)
for c in cars:
    if ((cars[c]['hp']/cars[c]['lbs'])) == largest:
        carWithlargest = c
print ("The car with largest power-to-weight ratio is : " + str(carWithlargest) +
       " the ratio is : %.8f" % largest)
```

### Extra credit (5 points)

When you look at the plot, it seems to suggest that as the power-to-weight ratio goes up, the time to 100 mph goes down, and this seems logical too. But we can go a bit further: we can compute a number, the <i>correlation coefficient</i> of two variables that has a value between -1 and 1. Correlation coefficients with an absolute value that is close to 1 indicate that the variables are linearly related by an equation of the form:
\begin{equation*}
y = mx + b
\end{equation*}
We won't go into the details of how to compute a correlation coefficient from scratch here. We'll just use a function<br> <code>pearsonr(power_to_weight_ratios, performance)</code> from the Python library <code>scipy.stats</code>. This function is documented at https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html. This function returns a 2-tuple whose first member is the correlation coefficient. Are our <code>performance</code> and<code>performance</code>? Is m positive or negative in the relationship?

```python
from scipy.stats import pearsonr
# Your code goes here to compute and print the correlation coefficient.
m = pearsonr(power_to_weight_ratios, time_to_100)
print(m)
```

### Extra Credit Answer
Yes our <code>time_to_100</code> and <code>performance</code> are lineary related because the absolute value of <code>m</code> which is our <code>correlation_coefficient</code> is close to 1. m is negative in this relationship as the slope goes from <code>up</code> to <code>down</code>
